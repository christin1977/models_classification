```python
# type d entreprise selon nombre de salariés
def determine_type_ent(enrollment):
    if enrollment < 3:
        return 'microentreprise'
    elif 3 <= enrollment <= 10:
        return 'TPE'
    elif 11 <= enrollment <= 250:
        return 'PME'
    elif 251 <= enrollment <= 5000:
        return 'ETI'
    else:  # enrollment > 2500
        return 'GE'
source_entreprise['type_ent'] = source_entreprise['enrollment'].apply(determine_type_ent)

# Convertir la colonne "creation_date" en type datetime / Extraire l'année et créer une nouvelle colonn
source_entreprise['creation_date'] = pd.to_datetime(source_entreprise['creation_date'])
source_entreprise['creation_year'] = source_entreprise['creation_date'].dt.year

# Supprimer la seconde colonne 'ape_code' dupliquée
source_entreprise = source_entreprise.loc[:, ~source_entreprise.columns.duplicated()]

# ratio moyen 2018-2022
source_entreprise['resultat_exploitation_ratio'] = source_entreprise[['resultat_exploitation_ratio_2021', 'resultat_exploitation_ratio_2020', 'resultat_exploitation_ratio_2019', 'resultat_exploitation_ratio_2018']].mean(axis=1)
source_entreprise['net_result'] = source_entreprise[['net_result_2021', 'net_result_2020', 'net_result_2019', 'net_result_2018']].mean(axis=1)
source_entreprise['current_assets'] = source_entreprise[['current_assets_2021', 'current_assets_2020', 'current_assets_2019', 'current_assets_2018']].mean(axis=1)
source_entreprise['equity'] = source_entreprise[['equity_2021', 'equity_2020', 'equity_2019', 'equity_2018']].mean(axis=1)
source_entreprise['current_liabilities'] = source_entreprise[['current_liabilities_2021', 'current_liabilities_2020', 'current_liabilities_2019', 'current_liabilities_2018']].mean(axis=1)
source_entreprise['revenue'] = source_entreprise[['revenue_2018', 'revenue_2019', 'revenue_2020', 'revenue_2021', 'revenue_2022']].mean(axis=1)
source_entreprise['robustness_ratio'] = source_entreprise[['robustness_ratio_2018', 'robustness_ratio_2019', 'robustness_ratio_2020', 'robustness_ratio_2021', 'robustness_ratio_2022']].mean(axis=1)
source_entreprise['solvability_ratio'] = source_entreprise[['solvability_ratio_2018', 'solvability_ratio_2019', 'solvability_ratio_2020', 'solvability_ratio_2021', 'solvability_ratio_2022']].mean(axis=1)
source_entreprise['net_debt'] = source_entreprise[['net_debt_2018', 'net_debt_2019', 'net_debt_2020', 'net_debt_2021', 'net_debt_2022']].mean(axis=1)
source_entreprise['gross_operating_ebitda'] = source_entreprise[['gross_operating_ebitda_2021', 'gross_operating_ebitda_2020', 'gross_operating_ebitda_2019', 'gross_operating_ebitda_2018']].mean(axis=1)
source_entreprise['gross_operating_ebitda_ratio'] = source_entreprise[['gross_operating_ebitda_ratio_2021', 'gross_operating_ebitda_ratio_2020', 'gross_operating_ebitda_ratio_2019', 'gross_operating_ebitda_ratio_2018']].mean(axis=1)
source_entreprise['CAGR_18_22'] = source_entreprise[['growth_ratio_2021', 'growth_ratio_2020', 'growth_ratio_2019']].mean(axis=1)

#hypothese taux interet de la dette à 3% / taux de rendement des actionnaires à 8%
source_entreprise['RCI_2018']=source_entreprise['net_result_2018']/(source_entreprise['current_assets_2018']-source_entreprise['current_liabilities_2018'])
source_entreprise['RCI_2019']=source_entreprise['net_result_2019']/(source_entreprise['current_assets_2019']-source_entreprise['current_liabilities_2019'])
source_entreprise['RCI_2020']=source_entreprise['net_result_2020']/(source_entreprise['current_assets_2020']-source_entreprise['current_liabilities_2020'])
source_entreprise['RCI_2021']=source_entreprise['net_result_2021']/(source_entreprise['current_assets_2021']-source_entreprise['current_liabilities_2021'])
source_entreprise['RCI_2022']=source_entreprise['net_result_2022']/(source_entreprise['current_assets_2022']-source_entreprise['current_liabilities_2022'])
source_entreprise['RCI']=source_entreprise['net_result']/(source_entreprise['current_assets']-source_entreprise['current_liabilities'])
source_entreprise['total_fonds']=source_entreprise['equity']+source_entreprise['net_debt']
source_entreprise['equity_costs']=(source_entreprise['equity']/source_entreprise['total_fonds']*0.08)+(source_entreprise['net_debt']/source_entreprise['total_fonds']*0.03)
source_entreprise['taux de recrutement']=(source_entreprise['indeed_job_results_count']/source_entreprise['enrollment'])
source_entreprise['RdtCpx_2018']=source_entreprise['net_result_2018']/(source_entreprise['equity_2018'])
source_entreprise['RdtCpx_2019']=source_entreprise['net_result_2019']/(source_entreprise['equity_2019'])
source_entreprise['RdtCpx_2020']=source_entreprise['net_result_2020']/(source_entreprise['equity_2020'])
source_entreprise['RdtCpx_2021']=source_entreprise['net_result_2021']/(source_entreprise['equity_2021'])
source_entreprise['RdtCpx_2022']=source_entreprise['net_result_2022']/(source_entreprise['equity_2022'])
source_entreprise['RdtCpx']=source_entreprise['net_result']/(source_entreprise['equity'])
source_entreprise['FRN_2018']=source_entreprise['current_assets_2018']-(source_entreprise['current_liabilities_2018'])
source_entreprise['FRN_2019']=source_entreprise['current_assets_2019']-(source_entreprise['current_liabilities_2019'])
source_entreprise['FRN_2020']=source_entreprise['current_assets_2020']-(source_entreprise['current_liabilities_2020'])
source_entreprise['FRN_2021']=source_entreprise['current_assets_2021']-(source_entreprise['current_liabilities_2021'])
source_entreprise['FRN_2022']=source_entreprise['current_assets_2022']-(source_entreprise['current_liabilities_2022'])
source_entreprise['FRN']=source_entreprise['current_assets']-(source_entreprise['current_liabilities'])
source_entreprise['reimbursement_ratio_2018']=source_entreprise['net_debt_2018']/(source_entreprise['equity_2018'])
source_entreprise['reimbursement_ratio_2019']=source_entreprise['net_debt_2019']/(source_entreprise['equity_2019'])
source_entreprise['reimbursement_ratio_2020']=source_entreprise['net_debt_2020']/(source_entreprise['equity_2020'])
source_entreprise['reimbursement_ratio_2021']=source_entreprise['net_debt_2021']/(source_entreprise['equity_2021'])
source_entreprise['reimbursement_ratio_2022']=source_entreprise['net_debt_2022']/(source_entreprise['equity_2022'])
source_entreprise['reimbursement_ratio']=source_entreprise['current_liabilities']/(source_entreprise['equity'])
```
